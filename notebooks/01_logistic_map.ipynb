```python
import sys
import os

# Agregamos el directorio raiz al path para poder importar src
sys.path.append(os.path.abspath('..'))

from src.systems.discrete import LogisticMap
from src.visualization.plot_styles import set_academic_style
from src.visualization.plot_utils import plot_trajectory

# Configuramos el estilo visual global
set_academic_style()
```

```python
# Caso 1: Convergencia a un punto fijo estable
# Teoria: Para 1 < r < 3, la poblacion se estabiliza en un valor fijo.

r_stable = 2.5
x0 = 0.1
steps = 50

# Instanciamos el modelo y simulamos
model_stable = LogisticMap(r=r_stable)
data_stable = model_stable.simulate(x0=x0, steps=steps)

# Visualizamos usando nuestra funcion limpia
plot_trajectory(data_stable, r=r_stable, title="Phase 1: Stable Fixed Point")
```

```python
# Caso 2: Duplicacion de periodo
# Teoria: Para 3 < r < 3.44, la poblacion oscila entre dos valores.

r_periodic = 3.2
model_periodic = LogisticMap(r=r_periodic)
data_periodic = model_periodic.simulate(x0=x0, steps=50)

plot_trajectory(data_periodic, r=r_periodic, title="Phase 1: Period Doubling Oscillation")
```

```python
# Caso 3: Regimen Caotico
# Teoria: Para r > 3.57, emerge el caos determinista.

r_chaos = 3.8
model_chaos = LogisticMap(r=r_chaos)
data_chaos = model_chaos.simulate(x0=x0, steps=100)

plot_trajectory(data_chaos, r=r_chaos, title="Phase 1: Deterministic Chaos")
```
